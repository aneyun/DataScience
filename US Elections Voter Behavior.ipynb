```r
#load libraries
library(ggplot2)
library(plyr)
library(dplyr)
library(reshape2)
library(scales)

cohens_d <- function(d1,d2){
    mu1 <- mean(d1)
    mu2 <- mean(d2)
    n1 <- length(d1)
    n2 <- length(d2)
    s1 <- sd(d1)
    s2 <- sd(d2)
    s <- sqrt((1/(n1+n2)) * ((n1-1)*s1*s1 + (n2-1)*s2*s2))
    d <- (mu1-mu2)/s
    return(d)
}
```

## Question 1: Do US voters have more respect for the police or for journalists?

### Introduce your topic briefly.  (5 points)
Explain how your variables are operationalized.  Comment on any gaps that you can identify between your operational definitions and the concepts you are trying to study.

Attitudes towards both police and journalists have been undergoing shifts in recent history. Events such as those in Ferguson, MO and the subsequent rise in the Black Lives Matter movement shine a spotlight on public perception of towards police. Similarly, President Trump's adversarial attitude toward journalists ("fake news"), shine a spotlight on public perception towards journalists.

Our approach to examining this question was to operationalize based on the variables available in the survey. There were a few relevant variables of interest for measuring respect towards police, for example, ftpolice, ftpolice_skp, ftpolice_page_timing, ftpolice_page, ord_ftpolice, honest, nonserious. 

We chose to go with ftpolice (and ftjournal) because of the directness of the survey question: 
*"How would you rate the police?"* 

Here, we will make an assumption that rating police highly implicitly means having more respect for the police and similarly for the journalists.

*ftpolice* has the following range of values: [0,100], where 100 implies maximum respect and 0 implies minimum respect.

*ftjournal* is somewhat similar to ftpolice except that the range of values is [-7,100]. -7 indicates that the question was skipped (when ftjournal_skp = 2) despite a request to non-response. We will limit the analysis to people who responded and scope the ftjournal variable range to [0,100].

### Perform an exploratory data analysis (EDA) of the relevant variables. (5 points)
This should include a treatment of non-response and other special codes, basic sanity checks, and a justification for any values that are removed.  Use visual tools to assess the relationship among your variables and comment on any features you find.

```r
data <- read.csv("anes_pilot_2018.csv", header = TRUE)
summary(data$ftjournal)

dataclean = data %>% filter(data$ftpolice>=0) %>% filter(data$ftjournal>=0)
paste('length of journal entries with all responses:' , length(data$ftjournal))

summary(dataclean$ftjournal)
paste('length of journal entries without non-response scores:' , length(dataclean$ftjournal))

g <- ggplot(dataclean, aes(x=ftpolice))
g <- g + geom_histogram(binwidth=10,fill = "grey", color = "black") + theme_bw() 
g + theme(legend.position = "bottom", legend.direction = "horizontal") + 
ggtitle("Histogram of responses to question about respect for police") +
labs(y="Count", x = "Rating for police")

g <- ggplot(dataclean, aes(x=ftjournal))
g <- g + geom_histogram(binwidth=10,fill = "grey", color = "black") + theme_bw() 
g + theme(legend.position = "bottom", legend.direction = "horizontal") +
ggtitle("Histogram of responses to question about respect for journalists") +
labs(y="Count", x = "Rating for journalists")

dat.m <- melt(dataclean,id.vars='caseid', measure.vars=c('ftpolice','ftjournal'))
p <- ggplot(dat.m, aes(variable,value)) +
      geom_boxplot() +
ggtitle("Comparison of responses to question about respect for police and journalists") +
labs(y="Rating", x = "Police (left), Journalists (right)")
p
```

### Based on your EDA, select an appropriate hypothesis test.  (5 points)
Explain why your test is the most appropriate choice, focusing on its statistical assumptions.

We considered the usage of a z or a t test. A z test needed a population s.d, which is not available. A paired t test was more appropriate as 
the assumptions for a t test are sastified:

* The dependent variable must be continuous (interval/ratio).
* The observations are independent of one another.
* The dependent variable should be approximately normally distributed.
* The dependent variable should not contain any outliers.


The dependent variable here is the difference of respect for police and respect for journalists. The test is paired because
we apply the t-test on the same sample with two different measures (police and journaists).


  0: 
Voters have equal respect for police and journalists. mu = 0

  :
Voters have more respect for police than for journalists mu > 0

Lastly, we perform a one-sided t-test because we would like to know which one of the two groups, police and journalists, is respected more. We assume a standard confidence interval of 95%.



### Conduct your test. (5 points)
Explain (1) the statistical significance of your result, and (2) the practical significance of your result.  Make sure you relate your findings to the original research question.

```r
#Statistical significance test:

t.test(dataclean$ftpolice , dataclean$ftjournal, mu = 0, alternative = "less", paired = TRUE, conf.level = .95 )
```

As the P-value is > .05, we fail to reject the null hypothesis. There is no statistically significant evidence to suggest that police are respected more than journalists. 

Practical significance: Cohens's d test helps us understand the practical significance of the t-test. Practical significance can be measured by the effect size. #Reference here: https://newonlinecourses.science.psu.edu/stat200/lesson/6/6.4

From the d value, we realize that the difference in means is not "Practically" significant because the d-value is only .4. A higher d value could have made the results practically significant. Thus, we conclude that our survey respondents had similar levels of respect for police and for journalists.

```r
#Practical Significance: Cohen's d
d <- cohens_d(dataclean$ftpolice, dataclean$ftjournal)
cat("Cohen's D:",d)
```

## Question 2: Are Republican voters older or younger than Democratic voters?

### Introduce your topic briefly.  (5 points)
Explain how your variables are operationalized.  Comment on any gaps that you can identify between your operational definitions and the concepts you are trying to study.

Politics seems to constantly be looking toward the future...the next election. The makeup of the future voter block is of intense interest. With the relatively large (by number) Millennial generation, a common reframe has been that "demographics is destiny", implying that a block of younger voters (even if they do not vote as frequently as older voters) who tend to be more liberal/Democrat will eventually create an unassailable majority.

For the purposes of evaluating this research question, we selected variables that come from the YouGov profile survey data. As defined in the codebook, these data come from previously-completed questionnaires.

Determining age is relatively straightforward. The survey data contains a variable "birthyr" that records the year in which each respondent was born. While it is possible that respondents did not answer this truthfully (we know from the survey responses that some small portion of respondents claimed they did not answer the survey questions honestly), we assume the quality control reviews conducted by YouGov (which rejected 131 responses) eliminated false answers and that therefore the answers given to this question are honest. An EDA conducted on the "birthyr" responses did not uncover any questionable results.

Determining party affiliation is a bit more problematic. We based the analysis on the variable "pid7x" which also comes from the profile survey data. For this variable, we used a response of 1 (Strong Dem) or 2 (Not very strong Dem) to indicate a Democrat voter, a response of 6 (Not very strong Rep) or 7 (Strong Rep) to indicate a Republican voter, and considered responses of 3, 4, or 5 to be Independents. We considered filtering these responses on only those who indicated they had voted in an election in 2018 (using variables "house18t", "senate18t", or "gov18t"), but decided that looking only at participation in one election year would not be representative of a "voter" as used by the research question. Multiple factors could have kept a "voter" from voting during this one particular election.

### Perform an exploratory data analysis (EDA) of the relevant variables. (5 points)
This should include a treatment of non-response and other special codes, basic sanity checks, and a justification for any values that are removed.  Use visual tools to assess the relationship among your variables and comment on any features you find.

```r
library(ggplot2)

data <- read.csv("anes_pilot_2018.csv", header = TRUE)

# Create new dataframe with only columns relevant for this research topic
voter_age <- data[,c('pid7x','birthyr')] #select(A, pid7x, birthyr)

voteridFunction <- function(x){
    if (x == 1 | x == 2) {
        id <- "Democrat"
    } else if(x == 3 | x == 4 | x == 5){
        id <- "Independent"
    } else if(x == 6 | x == 7){
        id <- "Republican"
    } else{
        id <- "Unknown"
    }
    return(id)
}

voter_age$partyid <- lapply(voter_age$pid7x, voteridFunction)
voter_age$partyid <- as.character(voter_age$partyid)
voter_age$partyid <- factor(voter_age$partyid)

current_year <- 2018
voter_age$age <- abs(voter_age$birthyr-current_year)

d_voter_age <- filter(voter_age, partyid == "Democrat")
r_voter_age <- filter(voter_age, partyid == "Republican")

cat("Summary for all parties:\n")
cat("Total number of entries for all parties", length(voter_age$birthyr))
summary(voter_age$age)

cat("Summary for Democrats:\n")
cat("Total number of entries for Democrats", length(d_voter_age$birthyr))
summary(d_voter_age$age)

cat("Summary for Republicans:\n")
cat("Total number of entries for Republicans", length(r_voter_age$birthyr))
summary(r_voter_age$age)

title <- "Histogram of age for Democratic voters"
g <- ggplot(d_voter_age, aes(x=age))
g <- g + geom_histogram(binwidth=3,fill = "grey", color = "black") + theme_bw() 
g + theme(legend.position = "bottom", legend.direction = "horizontal") + ggtitle(title)

title <- "Histogram of age for Republican voters"
g <- ggplot(r_voter_age, aes(x=age))
g <- g + geom_histogram(binwidth=3,fill = "grey", color = "black") + theme_bw() 
g + theme(legend.position = "bottom", legend.direction = "horizontal")+ ggtitle(title)


# boxplot x on y axis, left republicans, right democrats
g <- ggplot(voter_age, aes(partyid, age)) +
        geom_boxplot() +
    ggtitle("Boxplot of ages of voters by party") +
    labs(y="Age", x = "Party Identification")
g
```

The histogram for the age of Republican voters shows an approximately normal distribution. While the histogram for Democrat voters appears bi-modal (there appear to be two humps in the distribution), the sample size and CLT still allow us to satisfy an assumption of normal distribution. Finally, the boxplots show a visual confirmation of our hypothesis (stated below) that Republican voters are older than Democrat voters.

### Based on your EDA, select an appropriate hypothesis test.  (5 points)
Explain why your test is the most appropriate choice, focusing on its statistical assumptions.

The appropriate statistical test for this research question is a two sample two-tailed t-test. We will execute this test with a 95% confidence interval. This assumes:
* the data are continuous
* the data follow an approximately normal distribution (satisfied by the CLT and sample size and the robustness of the t-test)
* the variances of the two populations are equal
* the two samples are independent
* both samples are random samples from their respective populations

We suspect that Democrat voters are younger because a commonly held belief in American politics is that Republican voters tend to be older (and more conservative), while younger voters tend to be more liberal (and therefore likely to identify as Democrat). A two-tailed test is important, however, because while we believe this assumption and could set up our statistical test accordingly, others may not share this assumption and might become suspicious that we had started with a two-tailed test and then switched (thereby increasing our type I error rate). Two tailed is simply the more conservative approach.

### Conduct your test. (5 points)
Explain (1) the statistical significance of your result, and (2) the practical significance of your result.  Make sure you relate your findings to the original research question.

```r
# Run the t-test
t.test(x = d_voter_age$age, y = r_voter_age$age, mu = 0, alternative = "greater")

# Calculate Cohen's D to evaluate practical significance
d <- cohens_d(d_voter_age$age, r_voter_age$age)
cat("Cohen's D result is",d)
```

The results of the t-test indicate that we fail to reject the null hypothesis (that the means of the two groups are equal) in favor of the alternate hypothesis. Thus, there is no statistically significant difference in the mean age between Democrat and Republican voters, with the mean Democrat voter approximately 2.5 years younger.

On a practical level, 2.5 years does not seem terribly significant. The results of the Cohen's D test (0.16) are less than 0.2, so this test also indicates that the practical signifance is small.

While it is reasonable to believe that most people who believed that the Trump campaign coordinated with Russia (as asked in this question) would believe the investigation was not baseless, we felt that there could also be a fair number of people who did not believe coordination happened, but still believe Russia took actions on its own (and that the investigation was therefore justified).

Similarly, we considered the **mullerinv** variable for the question about approving or disapproving of the Mueller investigation, but rejected it as being too broad. We felt that there were numerous different reasons why one might disapprove of the Mueller investigation (e.g., the amount of time it took, its secrecy, the people in charge, Mueller himself, etc) even if one thought that the investigation was justified (not baseless).

While **russia16** did not precisely answer the research question, it was the **best fit**.

### Based on your EDA, select an appropriate hypothesis test.  (5 points)
Explain why your test is the most appropriate choice, focusing on its statistical assumptions.

The random variable here can be expressed as a Bernoulli RV where "1" depicts the case where the investigation is baseless, 
\begin{align*}
X &:= \begin{cases}
1,\; with\; probability\; P \\
0,\; with\; probability\; (1-P)
\end{cases}
\end{align*}
In the above equation, $P$ is used to define the hypothesis test. In this case the following hypothesis is formed.

$\textbf{ Null Hypothesis $H_0$:}$

Independent voters are equally likely to believe the investigation is baseless or not, i.e., 
$$P = 0.5$$

$\textbf{Alternate Hypothesis $H_a$:}$

Majority of independent voters believe the investigations to be baseless,
$$P > 0.5$$ 

*In this case, since the alternate hypothesis is trying to establish a probability greater than that of the null hypothesis, we will carry out a one-sided (right-sided) binomial test with a confidence of 95%*

### Conduct your test. (5 points)

**Statistical Significance:**

Since the underlying random variable is from a Bernoulli distribution, we need to carry out a one-sample binomial test for evaluating the Null hypothesis on the Independent voters (since the research question is specifically interested in the independent voters).


```r
n_baseless <- sum(data_q3[data_q3$party=='Independent',]$baseless)
n_independents <- length(data_q3[data_q3$party=='Independent',]$baseless)

binom.test(x=n_baseless,n=n_independents,p=0.5,alternative="greater")
```

In addition to the binomial test, a large-sample z-test is also appropriate, where both $X$ (random variable depicting the people who consider the investigation baseless) and the estimator $(\hat{p}= X/n)$ are approximately normally distributed. Therefore, we can compute the p-value based off the Z test statistic, using
\begin{align*}
Z &= \frac{\hat{p} - p_0}{\sqrt{p_0(1-p_0)/n}}
\end{align*}
and,
\begin{align*}
p_{value} &= 1 - \Phi(Z)
\end{align*}
where $\Phi$ is the standard normal distribution.

```r
#Alternate method (for large sample tests)
p_hat <- n_baseless/n_independents
p <- 0.5
q <- 1-p
np <- n_independents*p
nq <- n_independents*q

z <- (p_hat-p)/(sqrt(p*q/n_independents))
p_value <- (1-pnorm(z))
cat("p-value is " )
p_value
```

This is quite similar to the binomial test as should be expected.

$\textbf{Since the p-value is >>0.05, we fail to reject the null hypothesis.}$

Thus concluding, Independents are equally likely to feel that the investigations were baseless and not.

$\textbf{Practical Significance:}$

For a binomial test, the effect size or practical significance is estimated by computing the relative risk. Relative risk is defined as $\hat{p}/p_0$.

```r
p_hat <- n_baseless/n_independents
RR <- p_hat/0.5
cat("Effect size:",RR)
```

Since the relative risk is close to 1, the difference in the number of Independents, who think the investigations are baseless to that of those who think it is baseless, is not significant.

### Based on your EDA, select an appropriate hypothesis test.  (5 points)
Explain why your test is the most appropriate choice, focusing on its statistical assumptions.

**Null Hypothesis $H_0$:**

The effect of anger or fear in increasing the voter turnout is the same

**Alternate Hypothesis $H_a$:**

Effect of anger in increasing the voter turnout is not the same as fear

To answer this research question, we require a nonparametric hypothesis test because the variables of interest ("geangry" and "geafraid") represent responses on a likert scale. These are ordinal values, so a t-test would be inappropriate.

The Wilcoxon Signed Rank Test is the most appropriate hypothesis test. It requires the following assumptions:
* Data are paired and come from the same population. This is true in this sample, as the same respondent answered both questions.
* Each pair is chosen randomly and independently. This assumption holds for the respondents of the survey.
* Data are measured on an interval scale (and may be either the within pair differences or within pair comparison on an ordinal scale). From examining the question response options, this assumption is satisfied by the response options presented.
* Samples are dependent because they come from the same person. 

We select a Wilcoxon test with mu=0, asserting our hypothesis that there is no difference between anger and fear for voters who did not vote in 2016 and who did vote in 2018. We use a two-sided test and alternative hypothesis because we do not have reason to believe one outcome is more likely than another.

### Conduct your test. (5 points)
Explain (1) the statistical significance of your result, and (2) the practical significance of your result.  Make sure you relate your findings to the original research question.

```r
x <- as.numeric(data_q4[data_q4$new_voter==1,]$geafraid)
y <- as.numeric(data_q4[data_q4$new_voter==1,]$geangry)

wilcox.test(x,y ,alternative = "two.sided", paired = TRUE, exact = FALSE)

z <- 1.94 # taken from the standard normal table based on the outcome of the test
r <- z/(113^(1/2)) # 113 is the sample size
cat("Effect size is", r)
```

Because the p-value is (significantly) larger than 0.025, we cannot reject the null hypothesis. Calculating the effect size relies on the fact that as sample size grows the normal approximation method is used which allows us to calculate the z value based on the p-value calculated by the test. In this case, the effect size is less than 0.2, which is small-medium. The headline for this research question would be "Fear and anger are equally likely to have increased voter turnout between 2016 and 2018".
